```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#convert into ['Action','Adeventure','Fantasy','SciFi']

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
#ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Now doing the same thing for keywords columns
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#now here we taking the another column cast.Here we pickup only three imp cast 

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
##Now we are going for the crew column in which we take the job of director only
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        ps.stem(i)
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#vectorization=converting text into vectors.
#methods :Bag of Words

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,stop_words='english')
```

```python
#distance of individual movie to ither movie
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
#now we are going to create function for recommend movie
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
     
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        print(" ")
        print(new_df.iloc[i[0]].overview)
    
    return
```

```python
recommend("Avengers: Age of Ultron")
```
